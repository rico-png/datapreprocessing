```python
#importing our pandas library

import pandas as pd
```

```python
#reading our dataset

df = pd.read_csv("bank.csv", sep=";")
```

```python
#displaing the dataset

df.head()
```

```python
# Getting the categorical columns

cat_col = df.select_dtypes(['object']).columns
cat_col
```

```python
#converting categorical columns to numeric

df[cat_col] = df[cat_col].apply(lambda x:pd.factorize(x)[0])
df[cat_col] 
```

```python
#converting column to dummy

df_dummy =pd.get_dummies(df, columns=['job'])
df_dummy
```
